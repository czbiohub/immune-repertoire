```python
import glob
import pandas as pd

filenames = glob.iglob('GO_term_summary*.txt')
dfs = []
for filename in filenames:
    category = filename.split('_summary_')[-1].replace('-', ' ').split('.txt')[0]
    print(f'filename: {filename}\t\tcategory: {category}')
    df = pd.read_table(filename, index_col=None)
    df.index.name = df.columns[0]
    df.columns = list(df.columns[1:]) + ['dropme']
    df = df.drop(labels=['dropme'], axis=1)
    df['category'] = category

    dfs.append(df)
go_data = pd.concat(dfs)
print(go_data.shape)
go_data.head()
```

```python
go_data.to_csv('plasma_membrane_and_secretion_go_terms.csv')
```
